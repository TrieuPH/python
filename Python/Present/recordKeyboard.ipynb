```python
import keyboard
import mouse
import time

# Lists to store recorded events and time intervals
recorded_events = []
time_intervals = []

# Flags to indicate recording state
is_recording = True
last_timestamp = None

# Callback function to capture keyboard events
def on_key_event(e):
    global is_recording, last_timestamp
    
    if is_recording:
        current_timestamp = time.time()
        
        if last_timestamp is not None:
            time_intervals.append(current_timestamp - last_timestamp)
        
        last_timestamp = current_timestamp
        recorded_events.append((e.name, "keyboard", e.event_type, None))

# Callback function to capture mouse events
def on_mouse_event(e):
    global last_timestamp
    
    if is_recording:
        current_timestamp = time.time()
        
        if last_timestamp is not None:
            time_intervals.append(current_timestamp - last_timestamp)
        
        last_timestamp = current_timestamp
        if isinstance(e, mouse.ButtonEvent):
            recorded_events.append((e.button, "mouse", e.event_type, None, None))
        elif isinstance(e, mouse.MoveEvent):
            recorded_events.append(((e.x, e.y), "mouse", "move", None, None))

# Start recording keyboard events
keyboard.hook(on_key_event)

# Start recording mouse events
mouse.hook(on_mouse_event)

print("Recording keyboard and mouse actions. Press 'Shift + Enter' to stop recording.")

while is_recording:
    if keyboard.is_pressed("shift") and keyboard.is_pressed("enter"):
        is_recording = False

# Stop recording keyboard events
keyboard.unhook_all()

# Stop recording mouse events
mouse.unhook_all()

print("Recording stopped.")

# Save recorded events and time intervals to a Python file
with open('recorded_events.py', 'w') as file:
    file.write("recorded_events = [\n")
    for event, interval in zip(recorded_events, time_intervals):
        file.write(f"    {event},  # Time interval: {interval:.3f} seconds\n")
    file.write("]\n")

print("Recorded events saved to 'recorded_events.py'.")

# Generate automation script based on the recorded events and time intervals
automation_script = [
    "import pyautogui",
    "import mouse",
    "import time",
    "",
    "# Wait for 5 seconds before starting the automation",
    "time.sleep(5)",
    "# Start automation"
]

for event, interval in zip(recorded_events, time_intervals):
    event_info, event_source, event_type, *_ = event
    
    if event_source == "keyboard":
        if event_type == "down":
            automation_script.append(f"pyautogui.keyDown('{event_info}')")
        elif event_type == "up":
            automation_script.append(f"pyautogui.keyUp('{event_info}')")
            # Add time delay after key up event
            automation_script.append(f"time.sleep({interval:.3f})")
    elif event_source == "mouse":
        if event_type == "down":
            if isinstance(event_info, int):
                if event_info == mouse.LEFT:
                    automation_script.append("mouse.click(button='left')")
                elif event_info == mouse.RIGHT:
                    automation_script.append("mouse.click(button='right')")
                # Add time delay after mouse click event
                automation_script.append(f"time.sleep({interval:.3f})")
        elif event_type == "move":
            if event_info is not None:
                x, y = event_info
                # Add mouse move event with coordinates
                automation_script.append(f"mouse.move({x}, {y})")
                # Add time delay after mouse move event
                automation_script.append(f"time.sleep({interval:.3f})")


# Save automation script to a Python file
with open('automation_script.py', 'w') as file:
    for line in automation_script:
        file.write(line + "\n")

print("Automation script saved to 'automation_script.py'.")
```
