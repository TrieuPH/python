```python
#library imports
import pyautogui
import time
from pywinauto import Application, findwindows

def goInventory():
    # Wait for 5 seconds before starting the automation
    time.sleep(5)
    # Start automation
    pyautogui.keyDown('1')
    pyautogui.keyUp('1')
    time.sleep(0.1)
    pyautogui.keyDown('0')
    pyautogui.keyUp('0')
    time.sleep(0.1)
    pyautogui.keyDown('0')
    pyautogui.keyUp('0')
    time.sleep(0.1)
    pyautogui.keyDown('enter')
    pyautogui.keyUp('enter')
    time.sleep(0.1)
    pyautogui.keyDown('1')
    pyautogui.keyUp('1')
    time.sleep(0.1)
    pyautogui.keyDown('enter')
    pyautogui.keyUp('enter')
    time.sleep(0.1)
    pyautogui.keyDown('n')
    pyautogui.keyUp('n')
    time.sleep(2)
     # send arrow down 3 times
    for i in range(3):
        pyautogui.keyDown('down')
        pyautogui.keyUp('down')
        time.sleep(0.1)
    
    pyautogui.keyDown('4')
    pyautogui.keyUp('4')
    time.sleep(2)

def escPress():
    # Number of times to loop
    num_loops = 10
    # Loop through the key presses
    for _ in range(num_loops):
        # Press the Esc key
        pyautogui.keyDown('esc')
        time.sleep(0.1)
        
        # # Press N key
        # pyautogui.keyDown('n')
        # time.sleep(0.1)
        
        # Release the keys
        pyautogui.keyUp('n')
        pyautogui.keyUp('esc')

def focus_on_win_app(name):
    try:
        app = Application(backend="uia").connect(title_re=name)
        window = app.window(title_re=name)
        window.set_focus()
    except findwindows.ElementNotFoundError:
        pass
        return False
```

```python
## CREATE INVNETORY LOCATION ##
## This script will create a new inventory location in the system ##

location_name = 'showcase display 1'
location_description = 'showcase display 1'
location_name_create_test = 'showcase display test'
location_description_create_test = 'showcase display test'

def create_inventory(location_name, location_description):
    # Go to Inventory
    goInventory()
    # Send arrow down 5 times
    for i in range(5):
        pyautogui.keyDown('down')
        pyautogui.keyUp('down')
        time.sleep(0.1)
    # Go to Inventory Location
    pyautogui.keyDown('6')
    pyautogui.keyUp('6')
    time.sleep(3)
    pyautogui.keyDown('1')
    pyautogui.keyUp('1')
    time.sleep(3)
    pyautogui.keyDown('f2')
    pyautogui.keyUp('f2')
    time.sleep(1)
   
    
    # Send key location name
    pyautogui.write(location_name)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    # Send key location description
    pyautogui.write(location_description)

    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('down')
    pyautogui.keyUp('down')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('down')
    pyautogui.keyUp('down')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('tab')
    pyautogui.keyUp('tab')
    time.sleep(0.1)
    pyautogui.keyDown('f2')
    pyautogui.keyUp('f2')
    time.sleep(0.1)
    pyautogui.keyDown('enter')
    pyautogui.keyUp('enter')
    time.sleep(0.1)

    # Esc to BO login screen
    escPress()

focus_on_win_app("Oracle")
create_inventory(location_name_create_test, location_description_create_test)
```

```python
## ITEM TRANSFER ##
## This script is used to transfer items from one location to another ##

srcLocation = 'Default'
dstLocation = location_description
srcBucket = 'On Hand'
dstBucket = 'On Hand'
item = '1005'
quantity = 5

def item_transfer(srcLocation, dstLocation, srcBucket, dstBucket, item, quantity):
    goInventory()
    # send arrow down 4 times
    for i in range(4):
        pyautogui.press('down')
        time.sleep(0.1)
    # Select 5 to go to the item transfer menu
    pyautogui.press('5')
    # Delay for 1 second
    time.sleep(1)
    # Type the source location
    pyautogui.typewrite(srcLocation)
    # Press tab to go to the srcBucket
    pyautogui.press('tab')
    # Type the source bucket
    pyautogui.typewrite(srcBucket)
    # Press F8 to Progress
    pyautogui.press('f8')
    # Delay for 1 second
    time.sleep(1)
    # Type the destination location
    pyautogui.typewrite(dstLocation)
    # Press tab to go to the dstBucket
    pyautogui.press('tab')
    # Type the destination bucket
    pyautogui.typewrite(dstBucket)
    # Press F8 to Progress
    pyautogui.press('f8')
    # Select Reason Code, enter
    pyautogui.press('enter')
    # Delay for 1 second
    time.sleep(1)
    # Type the item
    pyautogui.typewrite(item)
    # Enter item
    pyautogui.press('enter')
    # Delay for 3 second
    time.sleep(3)
    # Press F3 change quantity
    pyautogui.press('f3')
    # Type the quantity
    pyautogui.typewrite(str(quantity))
    # Delay for 1 second
    time.sleep(1)
    # Press enter to confirm
    pyautogui.press('enter')
    # Press F2 to save
    pyautogui.press('f2')
    # Sleep for 2 second
    pyautogui.sleep(2)
    # Press Y to confirm
    pyautogui.press('y')
    # Press esc to back BO menu
    escPress()

focus_on_win_app("Oracle")
item_transfer(srcLocation, dstLocation, srcBucket, dstBucket, item, quantity)
```
