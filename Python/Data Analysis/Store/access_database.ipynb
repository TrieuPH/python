```python
import pandas as pd

# Assuming df is your original dataframe
df['invoiceDate'] = pd.to_datetime(df['invoiceDate'], dayfirst=True)
# Filter df for invoiceDate and VND conditions
#df_filter = df[(df['invoiceDate'] >= '2024-07-10 21:00:00') & (df['invoiceDate']<= '2024-07-11 23:00:00') & (df['VND'] > 0.0)]
df_filter = df[(df['invoiceDate'] >= '2024-10-25 07:00:00') & (df['invoiceDate'] <= '2024-10-25 23:00:00') & (df['VND'] > 0.0)]
df_filter
# Sort df_filter by invoiceDate
df_filter = df_filter.sort_values(by='invoiceDate')

# Check if there are any NaN values in the VND column
if df_filter['VND'].isnull().any():
    print('VND column has non-numeric value')
else:
    # Group by employeeName and calculate sum of VND for each group
    sum_VND_by_employee = df_filter.drop_duplicates(subset='invoiceCode').groupby('employeeName')['VND'].sum()

    # Iterate over each employee and print the sum of VND
    for employee, sum_VND in sum_VND_by_employee.items():
        # Format sum_VND to currency
        formatted_sum_VND = '{:,.0f}'.format(sum_VND)
        print(f'Sum of VND for {employee}: {formatted_sum_VND}')
    # Calculate total sum of VND
    total_sum_VND = sum_VND_by_employee.sum()
    print(f'Total sum of VND: {total_sum_VND:,.0f}')
```
