```python
import pandas as pd
import configparser
import os

# Read the value from the config.ini file
config = configparser.ConfigParser()
config.read('config.ini')
output_file_dir = config.get('path', 'output_file_dir')
output_file_name_net_sales = config.get('path', 'output_file_name_net_sales')
cateFileDir = config.get('path', 'cateFileDir')
# Read the category file
try:
    df_cate = pd.read_excel(cateFileDir)
except FileNotFoundError as e:
    df_cate = pd.read_excel(r"")
# using os.path.join() method
# output_file_dir = os.path.join(output_file_dir, 'outputFile')
net_sales_data = os.path.join(output_file_dir, output_file_name_net_sales) + '.csv'

# Read the csv file
# input_file should be the path to the merge csv file..
try:
    input_file = net_sales_data
    # read csv input file
    df = pd.read_csv(input_file)
except FileNotFoundError as e:
    input_file = r""
    df = pd.read_csv(input_file)
df.dropna(inplace=True)
```

```python
# Create new dataframe with InvoiceID include VN0071
df_VN0071 = df[df['InvoiceID'].str.contains('VN0071')]
df_VN0071
```

```python
# Add month column to df_VN0071_analysis from Date column
df_VN0071_analysis['Month'] = pd.DatetimeIndex(df_VN0071_analysis['Date']).month
df_VN0071_analysis
```

```python
# Show 10 Barcode best sales from each month in dataframe df_VN0071_analysis
# Create new dataframe with Barcode, Quantity, Month
df_VN0071_analysis_barcode = df_VN0071_analysis[['Barcode', 'Total Include VAT', 'Month']]
df_VN0071_analysis_barcode

# Show top 10 Barcode best sales from each month in dataframe df_VN0071_analysis_barcode
# Create pivot table with Barcode, Quantity, Month
df_VN0071_analysis_barcode_pivot = pd.pivot_table(df_VN0071_analysis_barcode, index=['Barcode'], values=['Total Include VAT'], columns=['Month'], aggfunc='sum')


# Add new column is sum Total Include VAT of 12 months
df_VN0071_analysis_barcode_pivot['Sum Total Include VAT'] = df_VN0071_analysis_barcode_pivot.sum(axis=1)
df_VN0071_analysis_barcode_pivot.sort_values(by=['Sum Total Include VAT'], inplace=True, ascending=False)
df_VN0071_analysis_barcode_pivot
```

```python
# Remove data with Sum Total Include VAT = 0 or NaN
df_VN0071_analysis_barcode_pivot = df_VN0071_analysis_barcode_pivot[df_VN0071_analysis_barcode_pivot['Sum Total Include VAT'] != 0]
df_VN0071_analysis_barcode_pivot.head(1000)
# filter = df_VN0071_analysis_barcode_pivot['Barcode'] == '8938502525351'
# Merge df_VN0071_analysis_barcode_pivot and df_cate
```
