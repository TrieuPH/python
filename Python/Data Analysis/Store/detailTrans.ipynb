```python
import os
import pandas as pd

try:
    data_path = r'C:\Users\PC\Dropbox\Projects\data_xls\detailTrans'
    files = os.listdir(data_path)
except:
    # If data path not exist
    data_path = '/Users/trieupham/Dropbox/Projects/data_xls/detailTrans'
    files = os.listdir(data_path)
dfs = []

for file in files:
    file_path = os.path.join(data_path, file)
    if file.endswith('.xlsx'):  # Ensure we're only processing Excel files
        try:
            # Attempt to read each Excel file
            df = pd.read_excel(file_path)
        except Exception as e:
            print(f'Error reading {file}: {e}')
        else:
            dfs.append(df)

combined_df = pd.concat(dfs, ignore_index=True)
print(combined_df)
combined_df.to_csv('combined_transactions.csv', index=False)
```

```python
# Function to sum "Tiền VND" for a specific employee on specific dates
def sum_tien_vnd(combined_df, employee_name, specific_dates):
    # Convert the invoiceDate to datetime format for filtering
    combined_df['invoiceDate'] = pd.to_datetime(combined_df['invoiceDate'], format='%d/%m/%Y %H:%M:%S')
    
    # Filter the DataFrame by the specific dates and employee name
    filtered_df = combined_df[
        (combined_df['invoiceDate'].dt.date.isin([pd.to_datetime(date).date() for date in specific_dates])) & 
        (combined_df['employeeName'].str.lower() == employee_name.lower())
    ]
    
    # Sum the "Tiền VND" column
    total_tien_vnd = filtered_df['Tiền VND'].sum()
    
    return total_tien_vnd

# Define the employee name and specific dates
employee_name = 'Nguyễn Ngọc Huỳnh Như'
specific_dates = ['20/06/2024', '']

# Calculate the total "Tiền VND" for the specified employee on the specified dates
total_tien_vnd = sum_tien_vnd(combined_df, employee_name, specific_dates)

# Change format of total_tien_vnd to xx.xxx.xxx
total_tien_vnd = '{:,.0f}'.format(total_tien_vnd)
print(f"Total 'Tiền VND' for {employee_name} on {', '.join(specific_dates)}: {total_tien_vnd}")
```
