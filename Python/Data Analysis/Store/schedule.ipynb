```python
import pandas as pd

# Use glob to find all CSV files in the folder
csv_files = glob.glob(os.path.join(folder_path, '*.csv'))

def calculate_hours(shift):
    normal_hours = 0
    night_hours = 0
    # Define night shift ranges
    night_shift_ranges = [("22:00", "24:00"), ("0:00", "6:00")]
    
    # Special handling for "19:00-23:00"
    if shift == "19:00-23:00":
        return 3, 1  # 3 normal hours, 1 night hour
    
    # Split the shift into start and end times
    start, end = shift.split("-")
    start_hour, start_minute = map(int, start.split(":"))
    end_hour, end_minute = map(int, end.split(":"))
    
    # Adjust for shifts that end after midnight
    if end_hour < start_hour:
        end_hour += 24
    
    # Convert times to minutes for easier comparison
    start_minutes = start_hour * 60 + start_minute
    end_minutes = end_hour * 60 + end_minute
    
    # Calculate total shift duration in minutes
    total_shift_minutes = end_minutes - start_minutes
    
    # Check each hour of the shift to determine if it's normal or night
    for minute in range(start_minutes, end_minutes):
        hour = (minute // 60) % 24  # Convert minute back to 24-hour format
        if (22 <= hour < 24) or (0 <= hour < 6):
            night_hours += 1
        else:
            normal_hours += 1
    
    # Convert minutes to hours
    normal_hours /= 60
    night_hours /= 60
    
    return normal_hours, night_hours

def process_shifts(file_path):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Add columns for total normal and night hours
    df['Total Normal Hours'] = 0.0
    df['Total Night Hours'] = 0.0

    # Iterate through each row and each date column
    for index, row in df.iterrows():
        for day in df.columns[1:]:  # Skip the first column (employee names)
            shifts = str(row[day]).split("/")
            for shift in shifts:
                if shift != "OFF" and "-" in shift:
                    normal_hours, night_hours = calculate_hours(shift)
                    df.at[index, 'Total Normal Hours'] += normal_hours
                    df.at[index, 'Total Night Hours'] += night_hours
    # Save the updated DataFrame to a new CSV file
    df.to_csv(file_path, index=False, encoding='utf-8-sig')
    # Return the updated DataFrame
    return df

csv_files = glob.glob(os.path.join(folder_path, '*.csv'))


for file_path in csv_files:
    print(f"Processing file: {file_path}")
    updated_df = process_shifts(file_path)
    # Optionally, save the updated DataFrame back to a file
    updated_df.to_csv(file_path, index=False, encoding='utf-8-sig')
```

```python
import pandas as pd
import glob
from datetime import datetime, timedelta

# Define the folder path and specific date

specific_date = '19/06/2024'

# Function to calculate hours from shift string
def calculate_hours(shift):
    if shift == 'OFF' or shift == '':
        return 0
    hours = 0
    for period in shift.split('/'):
        if '-' not in period:  # Check if the period contains a hyphen
            continue  # Skip this period if no hyphen is found
        start, end = period.split('-')
        # Adjust for '24:00' to be treated as '00:00' of the next day
        if end == '24:00':
            end = '00:00'
            end_dt = datetime.strptime(end, '%H:%M') + timedelta(days=1)
        else:
            end_dt = datetime.strptime(end, '%H:%M')
        start_dt = datetime.strptime(start, '%H:%M')
        # No need to check if end_dt < start_dt for shifts ending at '24:00' as it's handled above
        hours += (end_dt - start_dt).seconds / 3600
    return hours

# Use glob to find all CSV files in the folder
csv_files = glob.glob(f"{folder_path}/*.csv")

# Initialize the sum of total work hours
total_work_hours = 0

# Loop through each CSV file
for file_path in csv_files:
    df = pd.read_csv(file_path, encoding='ISO-8859-1')  # Assuming ISO-8859-1 encoding for simplicity
    
    # Check if the specific date column exists
    if specific_date in df.columns:
        # Calculate work hours for each row and sum
        total_work_hours += df[specific_date].apply(calculate_hours).sum()

# Print or return the total sum of work hours for the specific date
print(f"Total work hours on {specific_date}: {total_work_hours}")
```

```python
import pandas as pd
import glob
import os
from datetime import datetime, timedelta

def calculate_shift_hours(shift):
    if '-' not in shift:
        return 0
    start, end = shift.split('-')
    start_dt = datetime.strptime(start, '%H:%M')
    if end == '24:00':  # Handle '24:00' as '00:00' of the next day
        end_dt = datetime.strptime('00:00', '%H:%M') + timedelta(days=1)
    else:
        end_dt = datetime.strptime(end, '%H:%M')
    return (end_dt - start_dt).total_seconds() / 3600

def calculate_total_hours(folder_path, employee_name, start_date, end_date):
    total_hours = 0
    normalized_employee_name = employee_name.strip().lower()
    csv_files = glob.glob(os.path.join(folder_path, '*.csv'))
    for file_path in csv_files:
        df = pd.read_csv(file_path, encoding='utf-8')
        # Normalize column names and find the correct one for 'Họ và Tên'
        normalized_columns = {col.strip().lower(): col for col in df.columns}
        column_name = normalized_columns.get('họ và Tên', None)
        if column_name:
            # Normalize employee names in the DataFrame for comparison
            df[column_name] = df[column_name].str.strip().str.lower()
            df_filtered = df[df[column_name] == normalized_employee_name]
            for _, row in df_filtered.iterrows():
                for date_str in df.columns[1:]:  # Skip the first column
                    try:
                        # Adjust the date format if necessary
                        current_date = datetime.strptime(date_str.strip(), '%d/%m/%Y')
                        if start_date <= current_date <= end_date:
                            shifts = str(row[date_str]).split('/')
                            for shift in shifts:
                                total_hours += calculate_shift_hours(shift)
                    except ValueError:
                        continue  # Skip if the column name is not a date
        else:
            print(f"Column 'Họ và Tên' not found in {file_path}")
    return total_hours

# Define the folder path and employee name

employee_name = 'Huỳnh Thị Trà Mi'
start_date = datetime(2024, 6, 16)
end_date = datetime(2024, 6, 16)


# Calculate total work hours for June 23, 2024
total_hours = calculate_total_hours(folder_path, employee_name, start_date, end_date)
print(f"Total work hours for {employee_name} on June 23, 2024: {total_hours}")
```

```python
import pandas as pd
import glob
import os
from datetime import datetime, timedelta

# Define a list of the uploaded file paths
file_paths = [
    r'C:\Users\PC\Dropbox\Projects\github\Python\Python\Data Analysis\Store\Schedule\2_6-8_6.csv',
    r'C:\Users\PC\Dropbox\Projects\github\Python\Python\Data Analysis\Store\Schedule\9_6-15_6.csv',
    r'C:\Users\PC\Dropbox\Projects\github\Python\Python\Data Analysis\Store\Schedule\16_6-22_6.csv',
    r'C:\Users\PC\Dropbox\Projects\github\Python\Python\Data Analysis\Store\Schedule\23_6-25_6.csv',
    r'C:\Users\PC\Dropbox\Projects\github\Python\Python\Data Analysis\Store\Schedule\26_5-1_6.csv'
]

# Function to calculate the shift hours
def calculate_shift_hours(shift):
    if '-' not in shift:
        return 0
    start, end = shift.split('-')
    start_dt = datetime.strptime(start, '%H:%M')
    if end == '24:00':  # Handle '24:00' as '00:00' of the next day
        end_dt = datetime.strptime('00:00', '%H:%M') + timedelta(days=1)
    else:
        end_dt = datetime.strptime(end, '%H:%M')
    return (end_dt - start_dt).total_seconds() / 3600

# Function to calculate total Sunday hours
def calculate_sunday_hours(file_paths, employee_name):
    total_hours = 0
    normalized_employee_name = employee_name.strip().lower()
    start_date = datetime(2024, 1, 1)  # Example start date
    end_date = datetime(2024, 12, 31)  # Example end date
    current_date = start_date
    while current_date <= end_date:
        if current_date.weekday() == 6:  # Check if it's Sunday
            column_name = current_date.strftime('%d/%m/%Y').lstrip("0").replace("/0", "/")  # Removing leading zeros
            for file_path in file_paths:
                df = pd.read_csv(file_path, encoding='utf-8')
                normalized_columns = {col.strip().lower(): col for col in df.columns}
                actual_column_name = normalized_columns.get(column_name.strip().lower(), None)
                if actual_column_name and normalized_employee_name in df['Họ và Tên'].str.strip().str.lower().values:
                    df_filtered = df[df['Họ và Tên'].str.strip().str.lower() == normalized_employee_name]
                    for _, row in df_filtered.iterrows():
                        shifts = str(row[actual_column_name]).split('/')
                        for shift in shifts:
                            shift_hours = calculate_shift_hours(shift)
                            total_hours += shift_hours
                            print(f"Date: {column_name}, Shift: {shift}, Hours: {shift_hours}")
        current_date += timedelta(days=1)
    return total_hours

# Define the employee name
employee_name = 'Huỳnh Thị Trà Mi'

# Calculate total work hours on Sundays
total_sunday_hours = calculate_sunday_hours(file_paths, employee_name)
print(f"Total work hours for {employee_name} on Sundays: {total_sunday_hours}")
```
