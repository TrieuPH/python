```python
## iTru3.py for Python 3.6.5
# iTru3.py - a simple Python script to get the content of a website and export it to a text file
# iTru3.py - a simple Python script to get all links of a website and export them to a text file

# check if requests is installed, if not, install it
try:
    import requests
except ImportError:
    import os
    os.system('pip install requests')
    import requests
# get the content of a website

# url = 'https://www.vietstock.vn'
url = 'https://storefc.gs25.com.vn'
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
result = requests.get(url, headers=headers)
#print(result.content.decode())
# export result.content.decode() to text file
with open('iTru3_1.txt', 'w', encoding='utf-8') as f:
    f.write(result.content.decode())
# export all links to text file
# check if BeautifulSoup is installed, if not, install it
try:
    from bs4 import BeautifulSoup
except ImportError:
    import os
    os.system('pip install beautifulsoup4')
    from bs4 import BeautifulSoup

soup = BeautifulSoup(result.content.decode(), 'html.parser')
links = soup.find_all('a')
# for link in links:
#     print(link.get('href'))
# clean up text file, if there is any data, remove it
with open('iTru3_2.txt', 'w', encoding='utf-8') as f:
    f.write('')
# export all links to text file
with open('iTru3_2.txt', 'w', encoding='utf-8') as f:
    for link in links:
        href = link.get('href')
        if href is not None:
            f.write(href + '\n')
# Check if any duplicate links, remove them in the text file
with open('iTru3_2.txt', 'r', encoding='utf-8') as f:
    links = f.readlines()
    links = list(set(links))
with open('iTru3_2.txt', 'w', encoding='utf-8') as f:
    for link in links:
        f.write(link)

#print all links
for link in links:
    print(link)
```
