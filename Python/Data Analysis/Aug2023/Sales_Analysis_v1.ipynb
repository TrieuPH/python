```python
import pandas as pd
import xlsxwriter

# Load the list of StoreIDs from the Excel file
store_list_file = "C:\\Users\\minhtriet.pham\\Desktop\\New folder\\VscodeRun\\StoreListExport.xlsx"
store_id_list = pd.read_excel(store_list_file, header=0, usecols=[0])['StoreID'].tolist()

# Create a dictionary to store the sales contributions
sales_contributions = {}

# Iterate through unique StoreIDs and calculate sales contributions
for store_id in df_querystore2['StoreID'].unique():
    store_df = df_querystore2[df_querystore2['StoreID'] == store_id]
    contribution = store_df.groupby(['Month', 'Division Name']).agg({
        'Total Exclude VAT': 'sum',
        'Sum of Qty': 'sum'
    })
    sales_contributions[store_id] = contribution

# Create a Pandas Excel writer using XlsxWriter as the engine
excel_writer = pd.ExcelWriter('Sales_Contribution.xlsx', engine='xlsxwriter')

# Iterate through the sales contributions and write to Excel
for store_id, contribution in sales_contributions.items():
    
    # Calculate % Sales Contribution
    total_sales = contribution.groupby('Month')['Total Exclude VAT'].transform('sum')
    contribution['% Sales Contribution'] = (contribution['Total Exclude VAT'] / total_sales)
    
    # Rank the divisions for each month
    contribution['Rank'] = contribution.groupby('Month')['% Sales Contribution'].rank(ascending=False, method='first')
    
    contribution.to_excel(excel_writer, sheet_name=f'Store_{store_id}_Contribution')


# Close the Pandas Excel writer and save the file
excel_writer.close()

print("Sales contribution data exported to Sales_Contribution.xlsx")
```

```python
import openpyxl

# Load the Excel workbook
wb = openpyxl.load_workbook('Sales_Contribution.xlsx')

# Iterate through sheets and unmerge cells, then apply ffill
for sheet_name in wb.sheetnames:
    sheet = wb[sheet_name]
    
    # Create a copy of the merged cell ranges
    merged_cell_ranges = list(sheet.merged_cells.ranges)
    
    # Unmerge cells using the copied ranges
    for merged_cell_range in merged_cell_ranges:
        sheet.unmerge_cells(merged_cell_range.coord)
    
    # Apply ffill to fill the blank cells
    for column in sheet.iter_cols():
        for cell in column:
            if cell.value is None:
                cell.value = cell.offset(-1, 0).value

# Save the modified workbook
wb.save('Unmerged_Filled_Sales_Contribution.xlsx')
```

```python
import openpyxl
from openpyxl.styles import PatternFill
import pandas as pd

# Load the Excel workbook
wb = openpyxl.load_workbook('Unmerged_Filled_Sales_Contribution.xlsx')

# Create a Pandas DataFrame to store the summary data
summary_data = []

# Iterate through sheets and extract data for summary
for sheet_name in wb.sheetnames:
    if sheet_name.startswith("Store_"):
        store_id = sheet_name.split('_')[1]
        sheet = wb[sheet_name]
        
        # Iterate through rows and extract data
        for row in sheet.iter_rows(min_row=2, values_only=True):
            month = row[0]  # Column A is Month
            division = row[1]  # Column B is Division Name
            sales_contribution = row[4]  # Column E is % Sales Contribution
            summary_data.append([store_id, month, division, sales_contribution])

# Create a DataFrame from the summary data
summary_df = pd.DataFrame(summary_data, columns=['StoreID', 'Month', 'Division Name', '% Sales Contribution'])

# Add a new sheet for similar patterns
similar_patterns_sheet = wb.create_sheet(title='Similar Patterns', index=0)

# Write the header row to the similar patterns sheet
header_row = ['StoreID', 'Month', 'Pattern', 'Cumulative %']
unique_divisions = summary_df['Division Name'].unique()
for division in unique_divisions:
    header_row.append(division)
similar_patterns_sheet.append(header_row)

# Sort the DataFrame by StoreID, Month, and % Sales Contribution
summary_df.sort_values(by=['StoreID', 'Month', '% Sales Contribution'], ascending=[True, True, False], inplace=True)

# Define a list of distinct colors
distinct_colors = [
    "0070C0", "00B050", "FFC000", "ED7D31", "A5A5A5", "5B9BD5", "70AD47", "9E480E",
    "636363", "31859B", "4F81BD", "F79646", "C0504D", "4BACC6", "8064A2", "FABF8F",
    "C6E0B4", "A9D18E", "C27BA0", "BC95D6", "E7B6FF", "B4A7D6", "FFBB78", "5A9BD5"
]

# Create a dictionary to map divisions to colors
division_colors = {division: color for division, color in zip(unique_divisions, distinct_colors)}

# Iterate through unique combinations of Store ID and Month
for store_id, month in summary_df[['StoreID', 'Month']].drop_duplicates().itertuples(index=False):
    store_month_df = summary_df[(summary_df['StoreID'] == store_id) & (summary_df['Month'] == month)]
    cumulative_percentage = 0.0
    pattern = []
    division_contributions = {}
    for division, percentage in store_month_df[['Division Name', '% Sales Contribution']].itertuples(index=False):
        cumulative_percentage += percentage
        pattern.append(division)
        division_contributions[division] = percentage
        if cumulative_percentage >= 0.7:  # Check if cumulative % exceeds 70%
            break
    pattern_str = ' -> '.join(pattern)
    row_values = [store_id, month, pattern_str, cumulative_percentage]
    for division in unique_divisions:
        row_values.append(division_contributions.get(division, 0))
    similar_patterns_sheet.append(row_values)

# Apply percentage formatting to the Cumulative % and Division columns
for row in similar_patterns_sheet.iter_rows(min_row=2, min_col=4, max_col=similar_patterns_sheet.max_column):
    for cell in row:
        cell.number_format = '0.00%'
        
# Apply conditional formatting to highlight cells based on values in a row
for row in similar_patterns_sheet.iter_rows(min_row=2):
    values = [cell.value for cell in row[4:]]  # Start from the fifth column (Column E) to the end of the row
    non_zero_values = [value for value in values if value != 0]
    
    if non_zero_values:
        ranked_values = sorted(non_zero_values, reverse=True)
        
        for idx, cell in enumerate(row[4:], start=4):  # Start column index from 4 (Column E)
            value = cell.value
            if value != 0:
                rank = ranked_values.index(value)
                color_code = distinct_colors[rank % len(distinct_colors)]
                fill = PatternFill(start_color=color_code, end_color=color_code, fill_type="solid")
                cell.fill = fill

# Save the modified workbook
wb.save('Unmerged_Filled_Sales_Contribution.xlsx')

print("Similar Patterns sheet added to Unmerged_Filled_Sales_Contribution.xlsx")
```
