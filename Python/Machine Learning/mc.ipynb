```python
import spacy

nlp = spacy.load('en_core_web_sm')
doc = nlp(document_text)

for sentence in doc.sents:
    # Process each sentence and extract relevant information
    print(sentence.text)
```

```python
def generate_test_scenario(sentence):
    # Your logic to generate test scenario from sentence
    test_scenario = f"Test scenario: Verify {sentence}"
    return test_scenario

test_scenario_file = open('test_scenarios.txt', 'w')

for sentence in doc.sents:
    test_scenario = generate_test_scenario(sentence.text)
    test_scenario_file.write(test_scenario + '\n')

test_scenario_file.close()
```
